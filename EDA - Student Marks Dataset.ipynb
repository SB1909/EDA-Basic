```python
import pandas as pd
import matplotlib.pyplot as plt

#import matplotlib library for plotting some visuals.

#we have given data set in excel file format, so for the purpose of analysis we have to import it in python.
#We will do that by using pandas library.
#Imported data set will be in the form of DataFrame

data=pd.read_excel('Scores.xlsx')    
data
```

```python
#Accessing data frame column wise in the form of list:
# we need data in mutable form and we need to perfom indexing also 
#so list is best suitable type for extracting data from data frame.

names=list(data['Name'])
gender=list(data['Gender'])
city=list(data['CityTown'])
total_marks=list(data['Total'])
chem_marks=list(data['Chemistry'])
phys_marks=list(data['Physics'])
math_marks=list(data['Mathematics'])
```

#### Problem Statement:
What is the average mark in each subject achieved by male and female category?

#### Approach:
On gender basis we can calculate avg marks for each subject and we can compare them.

For comparision bar chart is best suitable.

So we are going to use bar chart for this and do the comparision between subjects and male and female category.

And find out who is performing well in specific subject and who's overall performance is better.
And from this we can conclude that which category need attention and to do well so they can perform better next time.

```python
#defining function to calculate avg mars for each subjects:

def avg_marks(marks):            
    gender=list(data['Gender'])    #extracting list of gender from data frame
    count_m=0
    count_f=0
    for i in gender:               #using for loop we will calculate no. of males and females present in given data
        if i=='M':
            count_m+=1
        elif i=='F':
            count_f+=1
    N=len(gender) 
    sum_total_male_mark=0
    sum_total_female_mark=0
    for i in range(len(gender)):   #using for loop we will calculate sum of total marks scored my each category 
        if gender[i]=='M':         #i will be iterate for total no. of males and females 
            sum_total_male_mark+=marks[i]    
        elif gender[i]=='F':       #if and elif statement for distributing marks as per male and females.
            sum_total_female_mark+=marks[i]
    avg_total_male_mark=sum_total_male_mark/count_m     #calculating avg marks
    avg_total_female_mark=sum_total_female_mark/count_f
    gender_total={}
    gender_total['Male']=avg_total_male_mark     #storing avg marks for thier Male or Female category in form of dictionary
    gender_total['Female']=avg_total_female_mark #so we can easily access it for further operation
    return gender_total
#taking input in the form of marks list of given subject:
avg_total_marks=avg_marks(total_marks)
avg_chem_marks=avg_marks(chem_marks)
avg_phys_marks=avg_marks(phys_marks)
avg_math_marks=avg_marks(math_marks)
```

```python
#plotting bar graph:
import numpy as np
gender1=list(avg_total_marks.keys())  #accessing gender list
total=list(avg_total_marks.values())  #accessing avg marks as per gender for all subjects in form of list
chem=list(avg_chem_marks.values())    #so we can give input to plot bar graph
phys=list(avg_phys_marks.values())
math=list(avg_math_marks.values())

fig=plt.figure(figsize=(10,5))     #definig plot size

x_axis=np.arange(len(gender1))     #defining x axis for all gender using numpy

#plotting bar graph for all subjects:
#(define position for plotting specific bar and its starting w.r.t x axis, mark list,label for which subject of marks it is, width of bar)
plt.bar(x_axis-0.3,total,label='Avg Total Marks',width=0.2)   
plt.bar(x_axis-0.1,math,label='Avg Mathematics Marks',width=0.2)
plt.bar(x_axis+0.1,chem,label='Avg Chemistry Marks',width=0.2)
plt.bar(x_axis+0.3,phys,label='Avg Physics Marks',width=0.2)



plt.xticks(x_axis,gender1)    # for x axis labels

plt.grid(True,'major','y')    #grid lines on y axis i.e. horizontal

plt.xlabel('Gender')          #defining labels for x and y axis
plt.ylabel('Avg Marks')  
plt.title('Avg marks for male and female category in each subject')  #giving title to plots
plt.legend()                  #plotting legend to show name of subject and color used for the same

plt.show       #finally ploting bar chart
```

#### Conclusions:
From above chart we can see comparision between average marks of subjects across male and female category.

1)Avg Marks in chemistry for female category is higher than for male category.

i.e. Females are doing well in chemistry. Male category needs some attention.

2)Avg marks in physics and mathematics is approx similar for both male and female.

i.e both genders have avg performance so no need of special attention to specific category in this subjects.

3)Overall performance of female candidates is better than male candidates.

#### Problem Statments:
What are the Maximum marks scored by student in each subject?
Who is the topper in each subject and who is the overall topper?

#### Approach:
To compare maximum marks scored by topper in given subjects we can use bar graph.

For this we need to calculate max marks in each subject and need to find out topper's name and sthen we have to plot bar graph using matplolib.

Then from conclusion we can get to know about topper's details like from which cities most of the toppers are and from what category either they are male or female.

```python
#definig function to calculate max marks:
def max_marks(sub):
    max_mark=max(data[sub])  #calculating max mark
    X=[]
    for i in range(len(data[sub])): #using for loop generting list of names of toppers
        if data[sub][i]==max_mark:
            x=data['Name'][i]
            X.append(x)
    return X,max_mark
#giving input as Subject for which we are finding topper name and their marks and storing both in seprate variable:
n_math,m_math=max_marks('Mathematics')
n_phys,m_phys=max_marks('Physics')
n_chem,m_chem=max_marks('Chemistry')
n_total,m_total=max_marks('Total')
```

```python
#plotting bar graph:
import numpy as np
subjects=data.columns[4:]      #creating list of subject names by accessing data frame column
max_marks=[m_math,m_phys,m_chem,m_total] #creating list os max marks in each subject..sequence is as per subjects given in data 
topper_names=[n_math,n_phys,n_chem,n_total] #creating list for topper name as per given sequence

fig=plt.figure(figsize=(15,6))   #defining plot size suitable
ax=plt.axes()
ax.set_facecolor('#FFFFE0')     #definigng background color for axes

#plotting bar graph:
plt.bar(subjects,max_marks,width=0.4,color='#000080')   #(x-axis var,y-axis var,width of bar,color for bar)

#using plt.text command we can plot toppers name on graphand thier marks also.
#used for loop to plot topper names for all subjects
for i in range(len(subjects)):
    plt.text(subjects[i],max_marks[i]+10,topper_names[i],ha='center',bbox=dict(figure=(1,1),fc='orange',alpha=0.9))   
    plt.text(subjects[i],max_marks[i]-9.5,max_marks[i],ha='center',color='white',fontsize=12)
    #(x-position,y-position,str=text to plot,ha== horizontal alignment=centeral,color==text color,text size)
    #bbox==we can define box around text in which we can define box size as figure, box color, aplha===transperancy)
    
plt.xlabel('Subjects')  #definig labels and title
plt.ylabel('Max Marks')  
plt.title('Max marks in each subject') 

plt.show    #printing graph
```

#### Problem Statement:

From 10 different cities student appeared for exam. By counting percentage of no. of students from different cities. We can visualize them in graphical format.
For showing percentage for different category, pie chart is best suitable. So using matplotlib in python we can visualize pie chart for the same.



```python
city_s=list(dict.fromkeys(city))  #making list all of cities present in data frame and removing duplicates also.
no_std_city=[]
for i in city_s:     #counting students from each city using for loop
    cn=city.count(i)
    no_std_city.append(cn)
prcnt_std=[]
for i in no_std_city:
    p=round((i/sum(no_std_city)*100),2)    #calculating percentage for no. of students from each city
    prcnt_std.append(p)
    
fig=plt.figure(figsize=(8,8))   

plt.pie(no_std_city,labels=prcnt_std)   #plotting pie chart

plt.title('Percentage students from each City')   
plt.legend(city_s)

plt.show  
```

```python
#defining functions for descriptive statistics:
def mean(Data):
    N=int(len(Data))
    summ=0
    for i in Data:
        summ=i+summ
    mean=summ/N
    return mean

def mode(Data):
    count_dic={}
    for i in Data:
        c=Data.count(i)
        count_dic[i]=c  #use dictionary to store count as velues for specific key 
    mode=max(count_dic,key=count_dic.get)  #getting max count from dictionary i.e. max values from dict
    return mode

def var(Data):
    N=int(len(Data))
    diff=[]
    for i in Data:
        diff_1=(i-mean(Data))**2
        diff.append(diff_1)
    summ_sq_diff=0
    for i in range(N):
        summ_sq_diff+=diff[i]
    var=summ_sq_diff/N
    return var

def std_dev(Data):
    std_dev=(var(Data))**0.5
    return std_dev

def rng_var_med(Data):
    r=max(Data)-min(Data)
    
    N=int(len(Data))
    diff=[]
    for i in Data:
        diff_1=(i-mean(Data))**2
        diff.append(diff_1)
    summ_sq_diff=0
    for i in range(N):
        summ_sq_diff+=diff[i]
    var=summ_sq_diff/N 
    
    Data.sort()
    N=int(len(Data))
    if N%2==0:
        median=(Data[int(N/2)-1]+Data[(int(N/2)+1)-1])/2
    else:
        median=Data[int((N+1)/2)-1]
    return r,var,median
```
